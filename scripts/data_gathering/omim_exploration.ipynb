```python
import pandas as pd
import json
```

```python
omim_genemap = pd.read_csv('../data/genemap2.txt',sep='\t',skiprows=3,skipfooter=64)
omim_genemap = omim_genemap[[
    '# Chromosome',
    'MIM Number',
    'Gene Symbols',
    'Approved Symbol',
    'Ensembl Gene ID',
    'Gene Name',
    'Phenotypes',
    'Comments'
]]
omim_genemap
```

```python
gpcr_names = pd.read_csv('../../../receptors/GuideToPharm_GPCR_targets_060121.csv')
gpcr_names = gpcr_names[['HGNC symbol','HGNC name']]
gpcr_names
```

```python
omim_gpcrs_by_symbol = gpcr_names.merge(omim_genemap,left_on='HGNC symbol',right_on='Approved Symbol',how='inner')
gpcr_omim_phenotypes = omim_gpcrs_by_symbol[['HGNC symbol','HGNC name','MIM Number','Phenotypes']][~omim_gpcrs_by_symbol.Phenotypes.isna()]
gpcr_omim_phenotypes
```

```python
gpcr_omim_phenotypes.to_csv('../data/gpcr_omim_phenotypes_060121.csv',index=False)
```
