```python
import pandas as pd
```

```python
pd.read_csv('../data/protein_sequences/GPCRdb_class_alignments/frizzled_alignment_GPCRdb_050321.csv',index_col=0,skiprows=2,header=None)
```

```python
residue_aas = (pd.read_csv('../data/protein_sequences/GPCRdb_class_alignments/frizzled_alignment_GPCRdb_050321.csv',index_col=0,skiprows=2,header=None)
         .drop(index='CONSENSUS')
         .melt(ignore_index=False)
         .reset_index())
residue_aas.columns = ['Protein','alignment_number','amino_acid']
residue_aas
```

```python
residue_numbers = pd.read_csv('../data/protein_sequences/GPCRdb_class_alignments/frizzled_alignment_GPCRdb_050321.csv',index_col=0,nrows=2,header=None).transpose().reset_index()
residue_numbers.columns = ['alignment_number','segment','GPCRdb_alignment_number']
residue_numbers['segment'] = residue_numbers['segment'].ffill()
residue_numbers
```

```python
residues = residue_aas.merge(residue_numbers,on='alignment_number')
residues[~residues.amino_acid.isin(['-','_'])].drop(columns='alignment_number')
```

```python
residues.to_csv('../data/protein_sequences/GPCRdb_generic_numbers_by_gpcr_residue_ClassF.csv')
```
