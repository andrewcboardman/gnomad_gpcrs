```python
clinvar_variants = pd.read_csv('../data/variant_summary.txt',sep='\t',low_memory=False)
clinvar_variants.head()
```

```python
clinvar_gpcr_variants = gpcrs.merge(clinvar_variants,left_on='HGNC Symbol',right_on='GeneSymbol')
clinvar_gpcr_variants = clinvar_gpcr_variants[[
    'HGNC Symbol',
    'Class',
    'Name',
    'ClinicalSignificance',
    'PhenotypeList',
    'OriginSimple',
    'ReviewStatus',
    'Assembly',
    'Chromosome',
    'PositionVCF',
    'ReferenceAlleleVCF',
    'AlternateAlleleVCF',
    'RS# (dbSNP)',
    'PhenotypeIDS'
]]
clinvar_gpcr_variants
```

```python
filtered_clinvar_classA_gpcr_variants = clinvar_gpcr_variants[
    (clinvar_gpcr_variants['Class']=='A') & \
    (clinvar_gpcr_variants['Assembly']=='GRCh37') & \
#    (clinvar_gpcr_variants['ReferenceAlleleVCF'].apply(len) == 1) & \
    (clinvar_gpcr_variants['OriginSimple']=='germline') & \
    (clinvar_gpcr_variants['ReviewStatus'] != 'no assertion criteria provided')
]
filtered_clinvar_classA_gpcr_variants
```
