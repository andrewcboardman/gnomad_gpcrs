```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3
import seaborn as sns
```

```python
gpcr_names = pd.read_csv('../../receptors/GuideToPharm_GPCR_targets_refined_060121.csv')
gpcr_names = gpcr_names[
    [
        'HGNC symbol',
        'HGNC name',
        'Human uniprot name'
    ]
]
gpcr_names
```

```python
schoeneberg = pd.read_csv('../data/gpcr_disease_genes_schoeneberg_050121.csv')
schoeneberg = schoeneberg.groupby(['HGNC symbol','Functional effect']).agg({'Schoeneberg disease associations': lambda x: x.tolist()}).reset_index()
schoeneberg
```

```python
omim = pd.read_csv('../data/gpcr_omim_phenotypes_060121.csv')
#mim['OMIM_Phenotypes'] = omim['OMIM_Phenotypes'].str.split(pat=';')
omim
```

```python
gpcrs_by_genetic_disease = gpcr_names.merge(omim,on='HGNC symbol',how='left')
gpcrs_by_genetic_disease = gpcrs_by_genetic_disease.merge(schoeneberg,on='HGNC symbol',how='left')
gpcrs_by_genetic_disease = gpcrs_by_genetic_disease.rename(columns={'Human uniprot name':'Uniprot name'})
gpcrs_by_genetic_disease = gpcrs_by_genetic_disease.pivot(
    index=['HGNC symbol','HGNC name','Uniprot name','OMIM Phenotypes'], 
    columns='Functional effect',
    values='Schoeneberg disease associations').reset_index()
gpcrs_by_genetic_disease = gpcrs_by_genetic_disease.loc[:,gpcrs_by_genetic_disease.columns.notnull()]
gpcrs_by_genetic_disease.columns = [
    'HGNC symbol',
    'HGNC name',
    'Uniprot name',
    'OMIM Phenotypes',
    'Gain of function diseases',
    'Loss of function diseases'
]
gpcrs_by_genetic_disease
```

```python
gpcrs_by_genetic_disease.columns
```

```python
drugs = pd.read_csv('../data/congreve2020_GPCR_targeted_drugs.csv')
drug_target_list_uniprot = drugs.groupby(['Uniprot name','MOA_simplified']).agg({'name':lambda x: x.tolist()}).reset_index()
drug_target_list_uniprot['Uniprot name'] = drug_target_list_uniprot['Uniprot name'].str.upper().str.slice(stop=-1)  + '_HUMAN'
drug_target_list_uniprot.columns = ['Uniprot name','Simplified MOA','Approved drugs']
drug_target_list_uniprot['Simplified MOA'] = drug_target_list_uniprot['Simplified MOA'].str.title()
drug_target_list_uniprot = drug_target_list_uniprot.pivot(index='Uniprot name',columns='Simplified MOA',values='Approved drugs').reset_index()
drug_target_list_uniprot.columns = ['Uniprot name','Activating approved drugs','Inactivating approved drugs']
drug_target_list_uniprot
```

```python
gpcr_diseases_and_drugs = gpcrs_by_genetic_disease.merge(drug_target_list_uniprot,on='Uniprot name', how='outer').drop(columns = ['HGNC symbol','HGNC name'])
gpcr_diseases_and_drugs.to_csv('../data/GPCR_phenotypes_and_drugs.csv')
gpcr_diseases_and_drugs
```

```python
gpcr_genes = pd.read_csv('../../receptors/Guide2Pharm_GPCR_mouse_homologs_070121.csv')
gpcr_genes = gpcr_genes[['HGNC symbol','HGNC name','MGI id','MGI symbol','MGI name']]
mouse_gpcr_genes = gpcr_genes[~gpcr_genes['MGI id'].isna()]
mouse_gpcr_genes
```

```python
impc_phenotype_hits = pd.read_csv('../data/impc_phenotypeHitsPerGene.csv.gz')
impc_phenotype_hits
```

```python
gpcr_mouse_phenotype_hits = mouse_gpcr_genes.merge(impc_phenotype_hits,how='left',left_on='MGI id',right_on='MGI Gene Id')
gpcr_mouse_phenotype_hits = gpcr_mouse_phenotype_hits[
    [
        'HGNC symbol',
        'MGI symbol',
        'MGI Gene Id',
        '# Phenotype Hits',
        'Phenotype Hits'
    ]
]
gpcr_mouse_phenotype_hits.columns = [
    'HGNC symbol',
    'MGI symbol',
    'MGI Gene Id',
    'Number of Mouse Phenotype Hits',
    'Mouse Phenotype Hits'
]
gpcr_mouse_phenotype_hits
```

```python
gpcr_mouse_phenotype_hits.to_csv('../data/GPCR_mouse_phenotype_hits.csv')
```

```python
gpcr_phenotypes_and_drugs['has_OMIM_Phenotypes'] = ~gpcr_phenotypes_and_drugs['OMIM Phenotypes'].isna()
gpcr_phenotypes_and_drugs['has_drugs'] = ~gpcr_phenotypes_and_drugs['Approved drugs'].isna()
gpcr_phenotypes_and_drugs['tested_mouse_phenotypes'] = ~gpcr_phenotypes_and_drugs['Number of Mouse Phenotype Hits'].isna()
gpcr_phenotypes_and_drugs['has_mouse_phenotypes'] = ~gpcr_phenotypes_and_drugs['Mouse Phenotype Hits'].isna()
gpcr_phenotypes_and_drugs[
    [
        'HGNC symbol',
        'HGNC name',
        'has_OMIM_Phenotypes',
        'has_drugs',
        'tested_mouse_phenotypes',
        'has_mouse_phenotypes'
    ]
]
```

```python
venn2(
    [
        set(gpcr_phenotypes_and_drugs[gpcr_phenotypes_and_drugs.has_OMIM_Phenotypes]['HGNC symbol'].values),
        set(gpcr_phenotypes_and_drugs[gpcr_phenotypes_and_drugs.has_drugs]['HGNC symbol'].values)
    ],
    set_labels = [
        'GPCRs linked to \n phenotypes in OMIM',
        'GPCRs targeted by \n FDA-approved drugs'
    ]
)
plt.tight_layout()
plt.savefig('../plots/drugs_omim_venn.png',dpi=200)
```

```python
venn3(
    [
        set(gpcr_phenotypes_and_drugs[
                gpcr_phenotypes_and_drugs.has_OMIM_Phenotypes & \
                gpcr_phenotypes_and_drugs.tested_mouse_phenotypes
        ]['HGNC symbol'].values),
        set(gpcr_phenotypes_and_drugs[
                gpcr_phenotypes_and_drugs.has_drugs & \
                gpcr_phenotypes_and_drugs.tested_mouse_phenotypes
        ]['HGNC symbol'].values),
        set(gpcr_phenotypes_and_drugs[
                gpcr_phenotypes_and_drugs.has_mouse_phenotypes
        ]['HGNC symbol'].values)           
    ],
    set_labels = [
        'GPCRs linked to \n phenotypes in OMIM',
        'GPCRs targeted by \n FDA-approved drugs',
        'GPCRs linked to \n phenotypes in knockout mice'
    ]
)
plt.tight_layout()
plt.savefig('../plots/Mouse_drugs_omim_venn.png',dpi=200)
```
