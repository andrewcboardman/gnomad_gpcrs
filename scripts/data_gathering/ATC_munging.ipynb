```python
import pandas as pd
```

```python
raw_atc_data = pd.read_csv('../data/drug_annotations/ATC.csv')
atc_data = raw_atc_data[['Class ID','Preferred Label','ATC LEVEL']].copy()
atc_data['is_ATC'] = atc_data['Class ID'].str.split('/').apply(lambda x: x[-2])
atc_data['ATC_code'] = atc_data['Class ID'].str.split('/').apply(lambda x: x[-1])
atc_data = atc_data[atc_data['is_ATC'] == 'UATC']
atc_data = atc_data[['ATC_code','ATC LEVEL','Preferred Label']]
atc_data.columns = ['ATC_code','ATC_level','Label_name']
atc_data
```

```python
drugs_by_atc = atc_data[atc_data['ATC_level']==5].copy().drop(columns='ATC_level')
drugs_by_atc['ATC_level_1'] = drugs_by_atc['ATC_code'].str.slice(stop=1)
# drugs_by_atc['ATC_level_2'] = drugs_by_atc['ATC_code'].str.slice(stop=3)
# drugs_by_atc['ATC_level_3'] = drugs_by_atc['ATC_code'].str.slice(stop=4)
drugs_by_atc = drugs_by_atc.merge(atc_data[atc_data['ATC_level']==1].drop(columns='ATC_level'),
                   left_on='ATC_level_1',right_on='ATC_code',suffixes=('','_level_1')).drop(columns='ATC_level_1')
drugs_by_atc
# drugs_by_atc = drugs_by_atc.merge(atc_data[atc_data['ATC_level']==2].drop(columns='ATC_level'),
#                    left_on='ATC_level_2',right_on='ATC_code',suffixes=('','_level_2'))
# drugs_by_atc.merge(atc_data[atc_data['ATC_level']==3].drop(columns='ATC_level'),
#                    left_on='ATC_level_3',right_on='ATC_code',suffixes=('','_level_3'))
```
