```python
import pandas as pd
```

```python
gpcr_genes = pd.read_csv('../../../receptors/GuideToPharm_GPCR_targets_060121.csv')
gpcr_genes = gpcr_genes[['HGNC symbol','HGNC name','MGI id','MGI symbol','MGI name']]
mouse_gpcr_genes = gpcr_genes[~gpcr_genes['MGI id'].isna()]
mouse_gpcr_genes
```

```python
impc_phenotype_hits = pd.read_csv('../data/phenotypeHitsPerGene.csv.gz')
impc_phenotype_hits.head()
```

```python
gpcr_mouse_phenotype_hits = mouse_gpcr_genes.merge(impc_phenotype_hits,how='inner',left_on='MGI id',right_on='MGI Gene Id')
gpcr_mouse_phenotype_hits[~gpcr_mouse_phenotype_hits['Phenotype Hits'].isna()]
```

```python
gpcr_mouse_phenotype_hits.to_csv('../data/GPCR_impc_phenotype_hits.csv')
```
