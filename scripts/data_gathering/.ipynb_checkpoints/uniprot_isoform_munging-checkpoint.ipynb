```python
import pandas as pd
```

```python
isoforms_by_ensembl_transcript = pd.read_csv('Uniprot_isoforms_by_ensembl_transcript.csv')
```

```python
isoforms_by_ensembl_transcript.head()
```

```python
isoforms_by_ensembl_transcript.columns = [
    'UniprotKB_ID',
    'UniprotKB_entry_name',
    'UniprotKB_status',
    'isoform_info',
    'Ensembl_transcripts'    
]
```

```python
isoforms_by_ensembl_transcript.head()
```

```python
isoforms_by_ensembl_transcript['Ensembl_transcripts'] = isoforms_by_ensembl_transcript['Ensembl_transcripts'].str.split(';')
```

```python
#isoforms_by_ensembl_transcript['Ensembl_transcripts']
ensembl_transcripts_by_isoform = isoforms_by_ensembl_transcript.explode('Ensembl_transcripts')
# = x.drop('isoform_info',axis=1)
ensembl_transcripts_by_isoform = ensembl_transcripts_by_isoform[ensembl_transcripts_by_isoform['Ensembl_transcripts'] != '']
```

```python
ensembl_transcripts_by_isoform
```

```python
ensembl_transcripts_by_isoform['UniprotKB_isoform'] = 'None'
```

```python
ensembl_transcripts_by_isoform.loc[~ensembl_transcripts_by_isoform['isoform_info'].isna(),'UniprotKB_isoform']
```

```python
ensembl_transcripts_by_isoform[~ensembl_transcripts_by_isoform['isoform_info'].isna()]['Ensembl_transcripts'].apply(lambda x: x.split('['))
```

```python
ensembl_transcripts_by_isoform[
    ~ensembl_transcripts_by_isoform['isoform_info'].isna()
]['Ensembl_transcripts'].apply(lambda x: x.split('[')).values
```
